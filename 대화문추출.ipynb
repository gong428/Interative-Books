```python
import re

# 파일 읽기
file_path = 'txt/LESMISERABLES_CH1.txt'

with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# 주인공 이름 (예: M. Myriel)
main_character = "M. Myriel"

# 큰따옴표와 작은따옴표로 감싸진 문장을 찾는 정규 표현식
quote_pattern = r'[“”"‘’\'](.+?)[“”"‘’\']'

# 정규 표현식을 사용해 대화문 추출
quotes = re.findall(quote_pattern, text)

# 주인공의 대사만 필터링하는 함수
def filter_main_character_dialogues(quotes, text, character_name):
    filtered_quotes = []
    previous_speaker = None
    for quote in quotes:
        # 대화문 앞뒤 문맥에서 주인공의 이름이 언급되었는지 확인
        context_pattern = rf'{character_name}'
        context_match = re.search(context_pattern, text, re.IGNORECASE)
        
        if context_match:
            # 문맥에서 주인공이 언급된 경우 주인공의 대사로 간주
            filtered_quotes.append(quote)
            previous_speaker = character_name
        else:
            # 문맥에 명시적인 언급이 없더라도 이전 대화자가 주인공이었다면 주인공의 대사로 간주
            if previous_speaker == character_name:
                filtered_quotes.append(quote)
            # 이후 다른 대화자가 나타나면 previous_speaker를 None으로 초기화
            else:
                previous_speaker = None
                
    return filtered_quotes

"""
문맥 검사:

대화문 앞이나 뒤에서 주인공의 이름이 언급되었는지 확인합니다.
주인공의 이름이 명시적으로 대화문 근처에 언급된 경우, 해당 대화문을 주인공의 대사로 간주합니다.

연속성 처리:

만약 대화문 앞뒤에 주인공의 이름이 명시적으로 언급되지 않았더라도, 이전에 주인공이 대화를 시작했다면 다음 대화문도 주인공의 것으로 간주할 수 있습니다.
이 방법은 대화문이 연속적으로 이어지는 경우에도 주인공의 대사를 추출하는 데 유용합니다.
"""

# 주인공의 대사만 필터링
main_character_quotes = filter_main_character_dialogues(quotes, text, main_character)

# 추출된 주인공의 대화문 출력
for idx, quote in enumerate(main_character_quotes, 1):
    print(f"Main Character Quote {idx}: {quote}")
```

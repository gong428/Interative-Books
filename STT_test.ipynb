```python
import gradio as gr
import whisper

# Whisper 모델 로드
model = whisper.load_model("base")

def transcribe_audio(audio_file):
    # Whisper 모델을 사용하여 음성을 텍스트로 변환
    result = model.transcribe(audio_file)
    recognized_text = result["text"]
    
    # 오디오 파일을 로드하고 로그 멜 스펙트로그램 생성
    audio = whisper.load_audio(audio_file)
    audio = whisper.pad_or_trim(audio)
    mel = whisper.log_mel_spectrogram(audio).to(model.device)
    
    # 언어 감지
    _, probs = model.detect_language(mel)
    detected_language = max(probs, key=probs.get)
    
    return recognized_text, detected_language

# Gradio 인터페이스 설정
iface = gr.Interface(
    fn=transcribe_audio, 
    inputs=gr.Audio(sources= "microphone",type="filepath"),  # 'source' 대신 'type'만 사용
    outputs=["text", "text"],
    live=False
    title="Whisper 음성 인식",
    description="마이크를 통해 입력된 음성을 텍스트로 변환하고, 감지된 언어를 출력합니다."
)

# 인터페이스 실행
iface.launch()
```

```python
import gradio as gr
import whisper

# Whisper 모델 로드
model = whisper.load_model("base")

def transcribe_audio(audio_file):
    # Whisper 모델을 사용하여 음성을 텍스트로 변환
    result = model.transcribe(audio_file)
    recognized_text = result["text"]
    
    # 오디오 파일을 로드하고 로그 멜 스펙트로그램 생성
    audio = whisper.load_audio(audio_file)
    audio = whisper.pad_or_trim(audio)
    mel = whisper.log_mel_spectrogram(audio).to(model.device)
    
    # 언어 감지
    _, probs = model.detect_language(mel)
    detected_language = max(probs, key=probs.get)
    
    # 음성 입력 후 상태를 None으로 반환하여 초기화
    return recognized_text, detected_language, None  # 마지막 출력에서 None으로 초기화

# Gradio 인터페이스 설정
iface = gr.Interface(
    fn=transcribe_audio, 
    inputs=gr.Audio(sources= "microphone",type="filepath"),  # 음성 입력 필드
    outputs=["text", "text"],  # 'state'를 사용하여 입력 필드를 초기화
    live=True,
    title="Whisper 음성 인식",
    description="마이크를 통해 입력된 음성을 텍스트로 변환하고, 감지된 언어를 출력합니다."
)

# 인터페이스 실행
iface.launch()
```
